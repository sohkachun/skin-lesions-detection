## Dependencies

```python
import platform
import subprocess
import sys

# Generate environment report
def generate_environment_report(output_file="environment_info.txt"):
    with open(output_file, "w") as f:
        # OS Information
        os_info = f"OS: {platform.system()} {platform.version()} (Release: {platform.release()})\n"
        f.write("# Environment Information\n\n")
        f.write("## Operating System\n")
        f.write(os_info + "\n")

        # Python Version
        python_version = sys.version
        f.write("## Python Version\n")
        f.write(python_version + "\n\n")

        # Installed Packages
        f.write("## Installed Python Packages\n")
        try:
            pip_freeze = subprocess.check_output([sys.executable, "-m", "pip", "freeze"]).decode("utf-8")
            f.write(pip_freeze + "\n")
        except subprocess.CalledProcessError as e:
            f.write(f"Error fetching Python packages: {e}\n")
    
    print(f"Environment report saved to {output_file}")

# Run the function
generate_environment_report()
```
